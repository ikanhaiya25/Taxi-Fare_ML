The features such as `store_and_fwd_flag` , `RatecodeID` , `payment_type` contain a large amount of NaN values.<br>
So we ned to check is there any relationship between these features and the target feature `total_amount`<br>
If not, we can remove them

Checking if the features have a string relationship with the target variable

```python
features = ['store_and_fwd_flag', 'RatecodeID', 'payment_type']

for feature in features:
    print(df[feature].unique())
```

```python
from scipy import stats
from scipy.stats import f_oneway
from scipy.stats import zscore

data = df[['store_and_fwd_flag','RatecodeID','payment_type','total_amount']].dropna()

# store_and_fwd_flag
group_n = data[data['store_and_fwd_flag'] == 'N']['total_amount']
group_y = data[data['store_and_fwd_flag'] == 'Y']['total_amount']

# RatecodeID
rate_groups = [
    data[data['RatecodeID']==i]['total_amount']
    for i in sorted(data['RatecodeID'].unique())
]

# payment_type
group_p1 = data[data['payment_type'] == 1]['total_amount']
group_p2 = data[data['payment_type'] == 2]['total_amount']

# t-testss
t1,p1 = stats.ttest_ind(group_n,group_y,equal_var=False)
t2,p2 = stats.ttest_ind(group_p1,group_p2,equal_var=False)

#Effect_Size
effect_size = (group_n.mean() - group_y.mean()) / data['total_amount'].std()
print("Effect size(store_and_fwd_flag):", effect_size)
effect_size_pt = (group_p1.mean() - group_p2.mean()) / data['total_amount'].std()
print("Effect size (payment_type):", effect_size_pt)

#Anova test(f-test)
f1,p3 = f_oneway(*rate_groups)

print(f"T1:{t1:.4f} P1:{p1:.4f}")
print(f"T2:{t2:.4f} P2:{p2:.4f}")
print(f"F1:{f1:.4f} P3:{p3:.4f}")

#null hypothesis
null_Hypothesis_SFF = "There is no difference in total_amount between store_and_fwd_flag 'N' and 'Y'"
null_Hypothesis_PT = "There is no difference in total_amount between payment_type '1' and '2'"

if p1 < 0.05:
    print(f"Reject null hypothesis: {null_Hypothesis_SFF}")
else:
    print(f"Fail to reject null hypothesis {null_Hypothesis_SFF}")

if p2 < 0.05:
    print(f"Reject null hypothesis: {null_Hypothesis_PT}")
else:
    print(f"Fail to reject null hypothesis {null_Hypothesis_PT}")


alpha = 0.05
if p3 < alpha:
    print("Reject null hypothesis: There is a relationship between RatecodeID and total_amount")
else:
    print("Fail to reject null hypothesis: There might not be a relationship between RatecodeID and total_amount")
```

For the feature `store_and_fwd_flag` and `payment_type`. t-test is used to determine if there is any relation between them and `total_amount`.<br>

For the feature 'RatecodeID'.Anova test(f-test) is performed to determind the relation between itself and `total_amount`.<br>

The P-value is 0 coz of the Large dataset.
Thats why effect_size shows how strong or meaning full the effect is

Effect size measures the magnitude of difference between groups.

# Feature Engineering

```python
#Converting the pickup and dropoff datetime columns to datetime type
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
#Converting the dtype of the features
df['fare_amount'] = df['fare_amount'].astype('int64')
df['payment_type'] = df['payment_type'].astype('int64')
```

```python
#Adding columns of pickup
df['pickup_hours'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day
df['pickup_weekday'] = df['tpep_pickup_datetime'].dt.weekday
df['pickup_month'] = df['tpep_pickup_datetime'].dt.month
```

It important to know the `pickup_hourse` , `pickup_day` , `pickup_weekday` , `pickupmonth` <br> Taxi Fare depend on the time patterns

```python
# Calculating trip duration
df['trip_duration(min)'] = ((df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds()/60).astype(int)
df.head()
```

## Reordering and Renaming Columns

```python
df = df[[
    'VendorID','passenger_count','trip_distance','RatecodeID','store_and_fwd_flag',
    'PULocationID','DOLocationID','payment_type','fare_amount','extra','mta_tax',
    'tip_amount','tolls_amount','improvement_surcharge','total_amount',
    'congestion_surcharge','trip_duration(min)'
    ]]

df.columns = [
    'vendor','passengers','distance','rate_id','flag','pickup_id','dropoff_id',
    'payment_type','fare','extras','tax','tip','tolls','improvement','total',
    'congestion','duration'
    ]

df['flag'] = df['flag'].map({'N':0,'Y':1})
df.head()
```

## Removing Duplicate values

```python
print(df.duplicated().sum())
```

```python
#dropping the duplicates and reseting the index
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)

#Cheching again after dropping the duplicates
print(df.duplicated().sum())
```

### Specifying numerical and categorical columns


```python
cat_cols = [
    'vendor','rate_id','flag',
    'pickup_id','dropoff_id','payment_type'
    ]

num_cols = [
    'passengers','distance','fare','extras','tax','tip','tolls',
    'improvement','congestion','duration'
    ]
```

### Standerdizing Data

```python
from sklearn.preprocessing import StandardScaler

#Standerd Scaling of numerical columns
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split

#Spliting data into feature and target
X = df[cat_cols + num_cols]
y = df['total']

#Split data into Traning and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, shuffle = False )
```

### Deciison Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Train the DecisionTreeRegressor model
dt_Model = DecisionTreeRegressor(random_state=42)
dt_Model.fit(X_train, y_train)

#Predict on the test set
y_pred = dt_Model.predict(X_test)

#Evaluate the model
mae_dt = mean_absolute_error(y_test, y_pred)
mse_dt = mean_squared_error(y_test, y_pred)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_test, y_pred)

#Printing the Values
print(f'Mean Absolute Error: {mae_dt}')
print(f'Mean Squared Error: {mse_dt}')
print(f'Root Mean Squared Error: {rmse_dt}')
print(f'R-squared: {r2_dt}')
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

#Train the RandomForestRegressor model
rfr_Model = RandomForestRegressor(
    n_estimators=20,
    max_depth=10,
    n_jobs=-1,
    random_state=42
)
rfr_Model.fit(X_train, y_train)

#Predict on the test set
y_pred = rfr_Model.predict(X_test)

#Evaluate the test set
mae_rfr = mean_absolute_error(y_test, y_pred)
mse_rfr = mean_squared_error(y_test, y_pred)
rmse_rfr = np.sqrt(mse_rfr)
r2_rfr = r2_score(y_test, y_pred)

#Printing the Values
print(f'Mean Absolute Error: {mae_rfr}')
print(f'Mean Squared Error: {mse_rfr}')
print(f'Root Mean Squaared Error: {rmse_rfr}')
print(f'R-squared: {r2_rfr}')
```

### Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

#Train the GradientBootingRegressor model
gbr_Model = GradientBoostingRegressor(
    n_estimators=50,
    max_depth=5,
    learning_rate=0.1,
    random_state=42
)
gbr_Model.fit(X_train, y_train)

#Predict on the test set
y_pred_gbr = gbr_Model.predict(X_test)

#Evaluating the test set
mae_gbr = mean_absolute_error(y_test, y_pred_gbr)
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
rmse_gbr = np.sqrt(mse_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

#Printing the Values
print(f'Mean Absolute Error: {mae_gbr}')
print(f'Mean Squared Error: {mse_gbr}')
print(f'Root Mean Squaared Error: {rmse_gbr}')
print(f'R-squared: {r2_gbr}')
```

```python
import xgboost as xgb

# Train the XGBoost Regressor model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)

# Predict on the test set
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate the model
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print("XGBoost Regressor Results:")
print(f'Mean Absolute Error: {mae_xgb}')
print(f'Mean Squared Error: {mse_xgb}')
print(f'Root Mean Squared Error: {rmse_xgb}')
print(f'R-squared: {r2_xgb}')
```
